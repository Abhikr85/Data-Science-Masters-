Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a popular and lightweight Python web framework used for developing web applications. 

Advantages of Flask Framework include:

1. Lightweight: Flask is a lightweight framework, which means it is easy to learn and does not have many dependencies. This makes it ideal for small to medium-sized web applications.

2. Flexibility: Flask is a flexible framework, which means it can be easily adapted to meet specific requirements. Developers can choose from a variety of extensions and libraries to add functionality to their applications.

3. Easy to get started: Flask has a simple and easy-to-use interface that makes it easy for developers to get started quickly.

4. Scalability: Flask can be used to develop applications of any size, from small personal projects to large enterprise-level applications.

5. Easy to test: Flask provides tools and libraries to make it easy to test web applications, which can help developers catch bugs and issues early in the development process.

6. Large community: Flask has a large and active community of developers who contribute to the framework, create extensions and libraries, and provide support to other developers.

Q3. What is App routing in Flask? Why do we use app routes?

In Flask, app routing refers to the process of mapping URLs to functions or methods that will handle the corresponding HTTP requests. App routing is a key aspect of building web applications in Flask, as it determines how incoming requests are processed and what response is returned to the client.

We use app routes in Flask for a few reasons:

1. Handling user requests: App routes allow us to define the endpoints or URLs that users can access in our application. For example, we can define a route to handle requests to the homepage, or a route to handle requests to a specific page on our site.

2. Separating concerns: App routes allow us to separate the concerns of handling different parts of our application. We can define different routes to handle different types of requests, and map them to different functions or methods that handle the corresponding logic.

3. Enabling RESTful APIs: App routes allow us to define RESTful APIs, which use HTTP methods like GET, POST, PUT, and DELETE to interact with data. By defining different routes for each API endpoint, we can create a clear and consistent interface for accessing and modifying data in our application.
